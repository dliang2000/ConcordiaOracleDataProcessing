```python
import json
import pandas as pd
```

```python
with open("../data.json", "r") as f:
    data = json.load(f)

json_list = []
for i in range(len(data)):
    numberOFRefactorings = len(data[i]["refactorings"])
    refactorings = data[i]["refactorings"]
    for j in range(numberOFRefactorings):
        if refactorings[j]["type"] == "Extract Method" and "test" in refactorings[j]["description"]:
            json_list.append(refactorings[j])

dfObj = pd.DataFrame(json_list)
dfObj
```
